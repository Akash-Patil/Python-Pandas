### The .query() Method

Method is used for filtering a df. Conditions:

1. Column names cannot have spaces in them
2. Conditions are strings based. No need to add the element(df) to condition


```python
bond.columns = [column_name.replace(' ', "_") for column_name in bond.columns]
bond.head(1)
```

```python
bond.query('Actor == "Sean Connery"') # filters df, see the usage of quotes for strings
bond.query("Actor != 'Roger Moore'").head()
```

```python
bond.query('Box_Office > 300 and Actor == "Roger Moore"') # use of the 'and' and 'or' permitted instead of symbols
```

```python
bond.query("Actor in ['Timothy Dalton', 'George Lazenby']") # can use 'in' and 'not in' operators as well
```

### .apply() Method on Single Columns

```python
bond = pd.read_csv('jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True) 
bond.head(3)
```

```python
def convert_to_string_add_millions(number): # custom function
    return str(number) + ' MILLIONS'

bond['Box Office'] = bond['Box Office'].apply(convert_to_string_add_millions) #applying func to main df
bond.head()
```

```python
# Alternate Bulk Logic/syntax

columns = ['Box Office', 'Budget', 'Bond Actor Salary']  # applying to multiple cols at once

for col in columns:
    bond[col] = bond[col].apply(convert_to_string_add_millions)
    
bond.head(3)
```

### The .apply() Method with Row values

```python
def good_movie(row):
    actor = row[1]  # filtering out the row values, [value] is for the column number
    budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return 'The best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'No info'
    
bond.apply(good_movie, axis = 'columns') # applying to rows, axis is columns as the values in rows are col associated
```

### The .copy() Method

Series by default act as views. The .copy() method creates a copy of the series so that we can make changes to series obj without affecting the original df. Modifications to copy object are isolated from main df and are done to the copy only.

```python
bond.head(3)
```

```python
directors = bond['Director'] 
directors.head(3)
```

```python
directors['A View to a Kill'] = 'Mister John Glen' # changing value in series acting as a view on df, affects the main df
```

```python
directors.head(3) # series value updated
```

```python
bond.head(3) # value in main df has been updated
```
