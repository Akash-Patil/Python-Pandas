```python
import pandas as pd
import numpy as np
```

### Changing Options with Attributes and Dot Syntax

```python
data = np.random.randint(0, 100, [1000, 50]) # generating random data, random numbers in a table
df = pd.DataFrame(data)
df.tail(2)
```

```python
pd.options.display.max_rows = 10 # truncate the display(from middle) 
df
```

```python
pd.options.display.max_columns = 10  # setting the max cols to be displayed -> will be applied through the notebook
df
```
